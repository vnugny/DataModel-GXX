# 🧪 Jinja Template Evaluator Test Notebook
This notebook loads a set of Jinja templates and evaluates them using the custom evaluator with date logic.

```
from jinja_template_evaluator_full import JinjaTemplateEvaluator
import json

evaluator = JinjaTemplateEvaluator()
```

```
with open("jinja_date_test_suite.json") as f:
    test_cases = json.load(f)

print(f"Loaded {len(test_cases)} test cases.")
```

```
for case in test_cases:
    print(f"\n### {case['description']} ###")
    try:
        result = evaluator.evaluate(case['template'], case.get('context', {}))
        print("Template: ", case['template'])
        print("Context: ", case.get('context', {}))
        print("Result:  ", result)
    except Exception as e:
        print("Error:", str(e))
```
